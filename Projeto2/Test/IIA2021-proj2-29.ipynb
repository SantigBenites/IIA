### Jogadores de ataque e de defesa obsessivos


Crie um dois jogadores obsessivos: o atacante obsessivo, prefere sempre por esta ordem, as acções: 'cima', 'esquerda','direita' e 'baixo'. O obsessivo da defesa prefere a célula mais no topo e mais à esquerda possível.
Realize um jogo entre eles e faça o seu display.

```python
def obsessive_attack(jogo,estado):
    movimentos = ["cima", "esquerda", "direita", "baixo"]
    for mov in movimentos:
        if estado.dir_action(mov) != "" :
            return mov    

def dist(p1, p2):
    return ((p1[0] - p2[0]) * (p1[0] - p2[0])) + ((p1[1] - p2[1]) * (p1[1] - p2[1]))

def obsessive_defense(game, state):
    #assume que existem acoes disponiveis
    acoes_disp = game.actions(state) # lista do estilo ["l,c", "l,c"]

    #distancia = sqrt((0 - l)^2 + (0 - c)^2), ou, de forma proporcional (0 - l)^2 + (0 - c)^2 -> que vamos minimizar
    acao_escolhida = acoes_disp[0]
    c_dist = dist((0,0), (int(acao_escolhida[0]), int(acao_escolhida[2])))
    for acao in acoes_disp:
        if dist((0,0), (int(acao[0]), int(acao[2]))) < c_dist:
            c_dist = dist((0,0), (int(acao[0]), int(acao[2])))
            acao_escolhida = acao

    return acao_escolhida
    

    
obsessive_game = Jogo2048_29({(3,0),(3,1)})
obsessive_game.jogar(obsessive_attack, obsessive_defense)

```

## Exemplos de jogadores alfabeta
 Descreva e teste nesta secção as várias funções de avaliação desenvolvidas tanto para o ataque como para a defesa.
 
Implementamos diversas funções de avaliação à medida que nos iamos familiarizando com o jogo, com as suas regras e táticas efetivas (curiosamente, acabaram por ser extremamente úteis na função de defesa).
 
 Inicialmente começamos por escrever funções de avaliação simples como o número de tiles vazios no board, o número de movimentos possíveis, a soma de todas as tiles do tabuleiro. No entanto estas rapidamente se demonstraram infrutríferas contra o hipolito. Mesmo a sua combinação gerava scores insatisfatórios.
 Após ler alguns artigos (e jogar o próprio jogo em demasia) verificamos existiam 3 "heurísticas" com maior relevância:
1. Smoothness - Avalia o quão "suave" é o board 
2. Monotonicity - Avalia se a board tem um layout "cascata", em que os tiles crescem ou decrescem em valor
3. Free Tiles - Como explicado acima, é o número de casas vazias

Aplicando estas funções de avaliação o score aumentou, mas não o suficiente para superar o hipolito...
No entanto à medida que iamos jogando tornou-se evidente que um padrão em S deixava o board menos vulnerável a que um 2 fosse colocado num "buraco" junto com tiles de valores superiores, o que é detrimental para o jogo.

Isto levou, finalmente, à derrota do hipolito. Não por cavalos, mas sim por cobras. No entanto estas vitórias não eram consistentes e por isso mais melhorias foram necessárias. Compensar as fraquezas da tática das cobras com as outras heurísticas como o número de 0's, a smoothness e a monotonicity ajudou a melhorar o score, mas apenas de forma incremental.
Ainda assim, a banha da cobra estava por terminar. Verificamos que por estar sempre a utilizar o mesmo padrão o algoritmo era demasiado inflexível. Por isso o que fizemos foi aplicar o mesmo padrão em todos os cantos, este algoritmo tornou-se mais flexível o que aumentou o score significativamente.

Após falarmos com o professor, expandimos os nossos horizontes e fomos introduzidos a mais alguns algoritmos, dos quais decidimos implementar; a procura de monte-carlo. Cuja ideia é excelente, no entanto ineficiente. Ainda assim serviu de inspiração para o algoritmo final.

Posto isto, o algoritmo que escolhemos foi: dado um estado, verifica as ações válidas e avalia cada uma delas através de uma combinação de heurísticas, a melhor destas é selecionada e o valor dessa avaliação é retornado.

No lado da defesa, num misto de sorte e engenho, a combinação do simetrico das heuristicas usadas no ataque em combinação com o a minimização do score (no caso, otilopih) obteve bons resultados.
